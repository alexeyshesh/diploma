```python
import os
import pandas as pd
from collections import Counter
```

```python
files = [f for f in os.listdir() if f[-3:] == 'txt']
```

```python
authors_list = sorted(list(dict(Counter([filename.split('_')[0] for filename in files])).keys()))
authors = {}
for i in range(len(authors_list)):
    authors[authors_list[i]] = i

authors
```

```python
from tqdm import tqdm

res = []

for name in tqdm(files):
    if not 'txt' in name:
        continue
    author = name.split('_')[0]
    with open(name, encoding='utf-8') as f:
        content = f.read()
        if len(content.split()) > 500:
            s = content.split()
            content_list = [s[i:i+500] for i in range(0, len(s), 500)]
            for elem in content_list:
                res.append([author, " ".join(elem)])
        else:
            res.append([author, " ".join(content.split())])
```

```python
df = pd.DataFrame(res, columns=['Author', 'Content'])
```
